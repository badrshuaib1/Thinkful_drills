```python
 ## loading the librarys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import re

```

```python
 csvfile = './welcome2013.csv'
df = pd.read_csv(csvfile, encoding ='Latin-1')
```

```python
## find out what needs to be done by checking the dataset and the variable that needs work done    

df=df.rename(columns = {'COST (£) charged to Wellcome (inc VAT when charged)':'Cost'})
df.head()
```

```python
## there are multiple ways a single publisher is inserted in the table.
## I will try and make each publisher has a single form.


df.Publisher.unique()
```

My plan is to abbriviate the names that has more than three words, and manage the rest of the names on their own

```python
df.Publisher = df.Publisher.str.strip()
```

```python
## in this part im taking out the most common differences. Later I will take out the most special parts on their own

## first ill try and make all the column capatalized and see how that changes the number of variations. 
df.Publisher = df.Publisher.str.upper()
df.Publisher = df.Publisher.str.replace('LTD', '')
df.Publisher = df.Publisher.str.replace('AND', '')
df.Publisher = df.Publisher.str.replace('INC', '')
df.Publisher = df.Publisher.str.replace('.', '')
df.Publisher = df.Publisher.str.replace('&', '')
df.Publisher = df.Publisher.str.replace('OF', '')
df.Publisher = df.Publisher.str.replace('FOR', '')
df.Publisher = df.Publisher.str.replace('-', '')
df.Publisher = df.Publisher.str.replace('THE', '')
df.Publisher = df.Publisher.str.replace('LLC', '')
df.Publisher = df.Publisher.str.strip()
```

```python
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\(.*\)', '',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\/.*', '',x))

```

```python
## In this part I will change the special cases that require specific changes, which wil require a mix of regexp and its methods. 


df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\bSOC\b', 'SOCIETY',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\bBIOLGY\b', 'BIOLOGY',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\bGENERMAL\b', 'GENERAL',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\bENDROCRINE\b', 'ENDOCRINE',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\bHAMATOLOGY|HEMATOLOGY\b', 'HAEMATOLOGY',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'ACS.*', 'ACS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'BMJ.*', 'BMJ',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'BENTHAM.*|BENTHAN.*', 'BENTHAM',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'WILEY.*|WLIEY.*', 'WILEY',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'WOLTERS.*', 'WOLTERS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'SPRINGER.*', 'SPRINGER',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'TAYLOR.*', 'TF',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'ASBM.*', 'ASBMB',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'CAMDUS.*|CADMUS.*', 'CADMUS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'CAMBRIDGE.*', 'CUP',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'COLD.*', 'CSHP',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'FEDERATION.*', 'FASEB',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'INMA.*', 'INMA HEALTHCARE',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'NEURO.*', 'NEUROSCIENCE',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'FRONTIERS.*', 'FRONTIERS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'SAGE.*', 'SAGE',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'MY.*', 'MY JOVE CORP',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'NATURE.*', 'NATURE PUBLISHING',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'OXD.*', 'OXD',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'PNAS.*', 'PNAS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'MIT.*', 'MIT PRESS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'BIOMED.*', 'BMC',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'DARMOUTH.*|DARTMOUTH.*', 'DARTMOUTH JOURNALS',x))   
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'HINDAWI.*', 'HINDAWI PUBLISHING',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'PUBMED.*', 'PUBMED',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'ROYAL.*|RSC.*', 'RSC',x))
df.Publisher = df.Publisher.str.replace('JOHN WILEY', 'WILEY')
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'\bBIOLOGIST\b|\bBIOLGISTS\b|\bBIOLOGISTSS\b', 'BIOLOGISTS',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'ELSEVIER.*', 'ELSEVIER',x))
df.Publisher = df.Publisher.apply(lambda x: re.sub(r'AMERICAN CHEMICAL.*', 'AMERICAN CHEMICAL SOCIETY',x))

```
