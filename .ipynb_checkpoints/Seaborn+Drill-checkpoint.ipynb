```python
response2 = req.get("https://api.census.gov/data/timeseries/healthins/sahie?get=NAME,NIC_PT,NUI_PT,PCTIC_PT,IPRCAT,IPR_DESC,AGECAT,AGE_DESC,SEXCAT,SEX_DESC,RACECAT,RACE_DESC&for=state:*&time=2016")


data2 = response2.json()
```

```python

res = req.get("https://api.census.gov/data/timeseries/healthins/sahie/variables.html")
soup = BeautifulSoup(res.content,'lxml')

table = soup.find_all('table')[0] 
df2 = pd.read_html(str(table))[0]

```

```python
ts1 = pd.DataFrame(data2)
new_header= ts1.iloc[0,]
ts1 = ts1.rename(columns = new_header)
ts1= ts1.iloc[1:,]
```

```python
ts1.head()
```

```python
ts1['PCTIC_PT'] = ts1['PCTIC_PT'].apply(float)
ts1['NIC_PT'] = ts1['NIC_PT'].apply(int) 
ts1['NUI_PT'] = ts1['NUI_PT'].apply(int)
ts1['total_of_population'] = ts1['NIC_PT'] + ts1['NUI_PT']
```

```python
df_income_poor = ts1.loc[lambda df: df['IPR_DESC'] == '<= 138% of Poverty', :]
df_income_middle= ts1.loc[lambda df: df['IPR_DESC'] == '138% to 400% of Poverty', :]
df_income_poor.head()
```

```python
df_income_4x = ts1.loc[lambda df: df['IPR_DESC'] == '<= 400% of Poverty', :]
df_income_total = ts1.loc[lambda df: df['IPR_DESC'] == 'All Incomes', :]


df_wealth_total = (df_income_total['total_of_population'].values - df_income_4x['total_of_population'].values)

df_wealth_insured = (df_income_total['NIC_PT'].values - df_income_4x['NIC_PT'].values)

df_wealth_perc = df_wealth_insured/df_wealth_total * 100
sex_cat = np.tile(('Both Sexes','Male Only','Female Only'), len(df_wealth_insured))
race_cat = np.tile(('All Races','All Races','All Races', 
                    'White alone, not Hispanic','White alone, not Hispanic',
                    'White alone, not Hispanic','Black alone, not Hispanic','Black alone, not Hispanic',
                    'Black alone, not Hispanic','Hispanic (any race)' ,
                    'Hispanic (any race)','Hispanic (any race)'), 
                len(df_wealth_insured))
df_wealth = pd.DataFrame({'NIC_PT': df_wealth_insured, 'PCTIC_PT': df_wealth_perc, 'SEX_DESC': sex_cat[:len(df_wealth_insured)], 'RACE_DESC':race_cat[:len(df_wealth_insured)],'IPR_DESC': np.repeat('> 400% of poverty', len(df_wealth_insured)) })

```

```python
df_wealth_race = df_wealth.loc[lambda df: (df['RACE_DESC'] != 'All Races') 
                                  & (df['SEX_DESC'] != 'Both Sexes'), :]
df_poor_race = df_income_poor.loc[lambda df: (df['RACE_DESC'] != 'All Races') 
                                  & (df['SEX_DESC'] != 'Both Sexes'), :]
```

```python


 #Setting the overall aesthetic

p1 = sns.FacetGrid(df_wealth_race, row="SEX_DESC", col="RACE_DESC", height=4, aspect=1)
# Name the plot type and the variable to be plotted using the structure.
p1.map(sns.distplot, "PCTIC_PT", color="steelblue")
# Moving the plots apart to make room for our titles.
plt.subplots_adjust(top=1, right = 2)
# Making a more informative axis name.
p1.set_axis_labels('Percentage Insured')


plt.show()


```

```python


p2 = sns.catplot(x='SEX_DESC',y='NIC_PT',hue='RACE_DESC',data=df_poor_race , kind="bar") 
p3= sns.catplot(x='SEX_DESC',y='PCTIC_PT',hue='RACE_DESC',data=df_wealth_race ) 
plt.title('Number Insured over Sex and Race')
p2.set(xlabel='', ylabel='Number Insured')


plt.show()
```

```python
p4= sns.distplot(df_poor_race["PCTIC_PT"])

plt.show()
```

```python
df_all_income=ts1.loc[lambda df: (df['RACE_DESC'] != 'All Races') & (df['SEX_DESC'] != 'Both Sexes') & (df['IPRCAT'] != '0') & (df['IPRCAT'] != '2')& (df['IPRCAT'] != '4') & (df['IPRCAT'] != '1'), :]
concatenated = pd.concat([df_all_income.assign(dataset='df_all_income'), df_wealth_race.assign(dataset='df_wealth_race')])
p5 = sns.violinplot(x='RACE_DESC',y='PCTIC_PT',hue='IPR_DESC',data=concatenated) 

plt.show()
```
